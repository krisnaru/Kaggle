```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
X = pd.read_csv("train.csv")
X.fillna(0, inplace=True)
```

```python
X.head()
```

```python
y = X.pop("SalePrice")
```

```python
numeric_variables = list(X.dtypes[X.dtypes != "object"].index)
X[numeric_variables].head()
```

```python
model = RandomForestRegressor(n_estimators=500, n_jobs=-1)
```

```python
model.fit(X[numeric_variables], y)
```

```python
X_test = pd.read_csv("test.csv")
X_test.fillna(X_test.mean(), inplace=True)
```

```python
train_pred = model.predict(X[numeric_variables])

print("Number of predictions:",len(train_pred))
```

```python
meanSquaredError=mean_squared_error(y, train_pred)
print("MSE:", meanSquaredError)
rootMeanSquaredError = sqrt(meanSquaredError)
print("RMSE:", rootMeanSquaredError)
```

```python
y_pred = model.predict(X_test[numeric_variables])
```

```python
submission = pd.DataFrame({
        "Id": X_test["Id"],
        "SalePrice": y_pred
    })
submission.to_csv('houseprices.csv', index=False)
```
